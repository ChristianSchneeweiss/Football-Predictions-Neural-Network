## Create single player array from player columns

```python
def player_concat_home(row):
    home_cols = ['home player 0','home player 1', 'home player 10', 'home player 11', 'home player 12',
                 'home player 13', 'home player 14', 'home player 2', 'home player 3',
                 'home player 4', 'home player 5', 'home player 6', 'home player 7',
                 'home player 8', 'home player 9',]
    home_players = [row[home_col] for home_col in home_cols if str(row[home_col]) != "nan"]
    return home_players

def player_concat_away(row):
    away_cols = ['away player 0', 'away player 1','away player 10', 
                 'away player 11', 'away player 12', 'away player 13',
                 'away player 2', 'away player 3', 'away player 4', 'away player 5',
                 'away player 6', 'away player 7', 'away player 8', 'away player 9']
    away_players = [row[away_col] for away_col in away_cols if str(row[away_col]) != "nan"]
    return away_players
```

```python
player_concat_away(football_df.iloc[0])
```

```python
football_df["home_players"] = football_df.apply(player_concat_home, axis=1)
football_df["away_players"] = football_df.apply(player_concat_away, axis=1)
```

## Create target label - Winner

```python
def create_target(row):
    if row["away goals"] > row["home goals"]:
        return "Away"
    elif row["away goals"] < row["home goals"]:
        return "Home"
    else:
        return "Draw"
    
football_df["winner"] = football_df.appl
```

```python
def remove_cols():
    keep_cols = ["home_players", "away_players", "away coach","away name","home coach","home name", "league",
                 "location", "referee", "season", "stadium"]
    return football_df[keep_cols]
clean_football_df = remove_cols()
clean_football_df.dropna(inplace=True)
```
